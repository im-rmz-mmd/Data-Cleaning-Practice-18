```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('stolen_vehicles.csv')
df
```

```python
df = df.drop(columns = ['vehicle_id', 'vehicle_desc'])
df = df.drop_duplicates()
df = df.dropna()
df
```

```python
df = df.astype({'make_id' : int, 'model_year' : int})
df['date_stolen'] = pd.to_datetime(df['date_stolen'])
df.dtypes
```

```python
df = df.to_excel('cleaned_theft.xlsx', index = False)
```
